```python
import json
import pandas as pd

from common.tokenizer import tokenizer
from tqdm import tqdm

tqdm.pandas()


with open("./database/train/minutes_train.json") as f:
    data = json.load(f)

tok = tokenizer("Kobart", "enc")

df = pd.DataFrame(data)[:10]
df["total_token"] = df.total.progress_apply(lambda x: tok._tokenize(x))
df["summary_token"] = df.summary.progress_apply(lambda x: tok._tokenize(x))

df.head()
```
